# **BREAST CANCER DETECTION**

## **PROBLEM STATEMENT**

**TO IDENTIFY WHETHER THE BREAST CANCER IS THE MALIGNANT TUMOR OR BENIGN TUMOR**

```python
import numpy as np                                                              # for using reshape function
import pandas as pd                                                             # for using dataframe

from sklearn.datasets import load_breast_cancer                                 # for loading the breast cancer dataset
from sklearn.preprocessing import StandardScaler                                # for converting the data range(-5, 5)

from keras.models import Sequential                                             # for creating sequential neural network
from keras.layers import Dense                                                  # for adding the layers to the neural network
```

## **DATA COLLECTION**

```python
data = load_breast_cancer()
data
```

```python
data.keys()
```

```python
for i in data.keys():
  print(i)
  print(data[i])
  print("---------------------------------------------------------------------------------------")
```

## **DATA PRE-PROCESSING**

```python
pd.DataFrame(data['data'], columns=data['feature_names'])
```

```python
data['data'] = StandardScaler().fit_transform(data['data'])                     # converting data into the range(-5, 5)
```

```python
df1 = pd.DataFrame(data['data'], columns=data['feature_names'])
df1
```

```python
df2 = pd.DataFrame(data['target'], columns=['labels'])
df2
```

```python
df = pd.concat([df1,df2], axis=1)                                               # combining dataframe df1 and df2
df
```

```python
df.sample(frac = 1)                                                             # shuffling the data
```

```python
features = df.values[:,:30]                                                     # getting first 30 columns for features
labels = df.values[:,30:]                                                       # getting the last column for labels
```

## **TRAINING THE MODEL**

```python
X_train = features[:500]                                                        # splitting 500 data for training
Y_train = labels[:500]

X_val = features[500:535]                                                       # splitting 35 data for validation
Y_val = labels[500:535]

X_test = features[535:]                                                         # splitting 34 data for testing
Y_test = labels[535:]
```

```python
model = Sequential()                                                            # creating the sequential neural network

model.add(Dense(15, activation='relu', input_dim=30))                           # adding the layers to the neural network
model.add(Dense(10,activation='relu'))
model.add(Dense(5,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])        # compiling the model to find the errors
```

```python
model.fit(X_train, Y_train, epochs = 10, batch_size = 5, validation_data=(X_val, Y_val)) # training the model using fit function
```

## **EVALUATING THE MODEL**

```python
model.evaluate(X_test, Y_test)                                                  # evaluating the model with test data
```

```python
model.evaluate(X_val, Y_val)                                                    # evaluating the model with validation data
```

```python
model.predict(np.reshape(X_test[0],(1,30)))                                     # reshaping the test data to (1,features) to predict the output
```

```python
Y_test[0]                                                                       # finding actual output
```

```python
for i in range(34):
  sample = X_test[i]
  sample = np.reshape(sample,(1,30))
  if model.predict(sample) > 0.5:
    print("predicted value = benign_tumor")
  else:
    print("predicted value = malignant_tumor")
  if Y_test[i] == 1:
    print("actual value = benign_tumor")
  else:
    print("actual value = malignant_tumor")
  print("---------------------------")
```
